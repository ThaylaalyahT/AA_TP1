```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from category_encoders import TargetEncoder
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
categorical_cols = ['brand', 'model', 'fuel_type', 'engine', 'transmission', 'ext_col', 'int_col', 'accident', 'clean_title']
 # numeric_cols = ['id', 'model_year', 'milage']

te = TargetEncoder(cols=categorical_cols)
train_data[categorical_cols] = te.fit_transform(train_data[categorical_cols], train_data['price'])
test_data[categorical_cols] = te.transform(test_data[categorical_cols])
```

```python
cols = train_data.columns 
features_names = cols.drop(['price'])
target = 'price'
```

```python
X_train = train_data[features_names]
Y_train = train_data[target]

X_test = test_data[features_names]
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
model
```

```python
Y_pred = model.predict(X_train)
Y_pred
```

```python
plt.scatter(Y_train, Y_pred)
plt.show()
```
